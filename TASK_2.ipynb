```python
# changing the columns to remove the empty spaces for data set 1
df.columns = ['State','Date','Frequency','Estimated Unemployment Rate (%)','Estimated Employed','Estimated Labour Participation Rate (%)','Area']
```

```python
#coverting the data into to datatime data set1
df.isetitem(df.columns.get_loc('Date'), pd.to_datetime(df['Date']))
df
```

**DATA INTERPRTATION**

```python
#State with highest Unemployment Rate
print(df['State'].value_counts().idxmax())

#State with Lowest Unemployment Rate
print(df['State'].value_counts().idxmin())

#Area with the Highest Unemployment Rate
print(df['Area'].value_counts().idxmax())

#Area with the Lowest Unemployment Rate
print(df['Area'].value_counts().idxmin())

```

```python
# adding month number and month name and year columns for analysis

#adding the month_number
df.loc[:, 'Month_num'] = pd.to_datetime(df['Date']).dt.month
# adding the month name
df.loc[:,'Month_name'] = pd.to_datetime(df['Date']).dt.month_name()
# adding the Year 
df.loc[:,'Year'] = pd.to_datetime(df['Date']).dt.year
df
```

```python
#Month with Highest Unemployment Rate
print(df['Month_name'].value_counts().idxmax())

#.Month with Lowest Unemployment Rate
print(df['Month_name'].value_counts().idxmin())

#.Year with Highest Unemployment Rate either 2019 or 2020
print(df['Year'].value_counts().idxmax())

```

```python
#.State with highest Unemployment Rate
print(df1['State'].value_counts().idxmax())

#State with Lowest Unemployment Rate
print(df1['State'].value_counts().idxmin())

#.Region with the Highest Unemployment Rate
print(df1['Region'].value_counts().idxmax())
#.Region with the Lowest Unemployment Rate
print(df1['Region'].value_counts().idxmin())
```

**DATA VISUALISATION**

```python
# TOP 15 STATES WITH HIGHEST UNEMPLOYMENT RATE
#data1 = df[['States','Estimated Unemployment Rate(%)']].groupby('state').sum().sort_values(by='Estimated Unemployment Rate(%)',ascending=False)
data1 = df[['State','Estimated Unemployment Rate (%)']].groupby('State').sum().sort_values(by='Estimated Unemployment Rate (%)',ascending=False)

data1.head(15)
```

```python
plot = px.bar(data1, x=data1.index, y="Estimated Unemployment Rate (%)", title = 'Top 15 States with Highest Unemployment Rate',color=data1.index) 
plot.show()
```

```python
#Pie Chart
plot = px.pie(df,values='Estimated Unemployment Rate (%)',names = df['State'],title="UNEMPLOYMENT RATE IN YEAR 2019-2020(state wise)")
plot.show()
plot = px.pie(df,values='Estimated Unemployment Rate (%)',names= df['Month_name'],title="UNEMPLOYMENT RATE IN YEAR 2019-2020(Monthwise)")
plot.show()
```

```python
#Unemployment Rate based on months
data2 = df[['Month_name','Estimated Unemployment Rate (%)']].groupby('Month_name').sum().sort_values(by='Estimated Unemployment Rate (%)',ascending=False)
data2
```

```python
plot = px.bar(data2,x=data2.index,y='Estimated Unemployment Rate (%)',title ='Top months of Highest Unemployment Rate',color=data2.index)
plot.show()
```

```python
#estimated unemployment rate year month and area
data3 = df[["Month_name","Year","Area","Estimated Unemployment Rate (%)"]].groupby(['Year', "Month_name",'Area']).mean()
#data3 = df[["Month_name","Year","Area","Estimated Unemployment Rate (%)"]].groupby(["Year","Month_name","Area"]).mean()
data3
```

**data visualisation**

```python
#Correlation matrix
cor = df.corr(numeric_only = True)
sns.heatmap(cor,annot=True)
```

```python
#Histograms
#fig = plt.figure(figsize = (10, 10))
sns.histplot(x='Estimated Unemployment Rate (%)', data=df, kde=True, hue='Area')
plt.title('Unemployment according to Area')
plt.xlabel('Unemployment Rate')
plt.show()
```

```python
#fig = plt.figure(figsize = (5, 5))
sns.lineplot(y='Estimated Unemployment Rate (%)', x='Date', data=df)
plt.title('Unemployment according to Date')
plt.xlabel('Date')
plt.xticks(rotation=90)
plt.ylabel('Unemployment Rate')
plt.show()
```

```python
#fig = plt.figure(figsize = (30, 15))
sns.histplot(x='Estimated Labour Participation Rate (%)', data=df, kde=True, hue='Area')
plt.title('Labour Participation according to Area')
plt.xlabel('Labour Participation Rate')
plt.show()
```

```python
#fig = plt.figure(figsize = (7, 7))
sns.lineplot(y='Estimated Labour Participation Rate (%)', x='Date', data=df)
plt.title('Labour Participation according to Date')
plt.xlabel('Date')
plt.xticks(rotation=90)
plt.ylabel('Labour Participation Rate')
plt.show()
```
